```python
x= np.arange(5)
y=[1,2,1,3,5]
p2 = np.polyfit(x,y,2)
p4 = np.polyfit(x,y,4)
```

```python
# Plot the polynomials and points
xx= np.linspace(-1,5,200)
plot(xx,np.polyval(p2,xx),label = 'Degree 2')
plot(xx,np.polyval(p4,xx), label = "Degree 4")
plot(x,y,'bo', markersize=2)
## Setting axis and legend
axis([-1,5,0,6])
legend(loc='upper left')
title('Polynomial fitting')
```

### Scatter Plots

```python
x1 = 2*np.random.standard_normal((2,100))
x2 = 0.8*np.random.standard_normal((2,100))+ np.array([[6],[2]])
plot(x1[0],x2[1],'b*',markersize = 6)
plot(x2[0],x1[1],'bo',markersize = 8)
title('2D Scatter Plot')
```

```python
## Logarithmic plots using loglog
x = np.linspace(0,10,200)
loglog(x,2*x**2,'b',linewidth=2, label = 'Quadratic polynomial')
loglog(x,4*x**4,'b_',linewidth = 2, label = "4th degree polynomial")
loglog(x,5*np.exp(x),'b:', linewidth=2,label = 'Exponential function')
title('Logarithmic plots')
legend(loc='upper left')
```

```python
#Plotting Histograms
sigma, mu=2,10
x = sigma*np.random.standard_normal(10000)+mu
hist(x,50)
z = np.linspace(0,20,200)
plot(z,(1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(z-mu)**2/(2*sigma**2)),'r',linewidth=3)
title=('Histogram with $\mu=[] and \quad* \sigma=[] $'.format(mu,sigma))
```

```python
# Using subplots to show several plots in one figure
def avg(x):
    """Simple Average"""
    return(roll(x,1)+x+roll(x,-1))/3
x = np.linspace(-2*np.pi,2*np.pi,200)
y = np.sin(x)+0.4+np.random.rand(200)
#Make succesive
for iteration in range (3):
    subplot(3,1,iteration +1)
    plot(x,y)
    yticks([])
    title('After (:d) average(s)'.format(iteration))
y= avg(y)
```

```python
def mandelbrot(h,w,maxit=20):
    """Returns image of Mandelbrot fractal of size h,w"""
    x=np.linspace(-2,0.8,w)
    y=np.linspace(-1.4,1.4,h)
    X,Y= np.meshgrid(x,y)
    c=X+Y*1j
    z=c
    divtime=maxit+ np.zeros(z.shape,dtype=int)
    for i in np.arange(maxit):
        z=z**2 +c
        diverge=z*np.conj(z)>z**2 #Diverging
        div_now = diverge & (divtime== maxit)
        divtime[div_now] = i
        z[diverge]=2
        return divtime
imshow(mandelbrot(400,400), cmap='seismic')
axis('off')
```
